```python
import seaborn as sns
```

```python
df = sns.load_dataset("tips")
```

```python
df.head()
```

```python
df.info()
```

### Correlation with Heatmap
##### A correlation heatmap uses colored cells, typically in a monochromatic scale, to show a 2D correlation matrix(table) between two disicrete dimensions or event types. It is very important in Feature Selection

```python
df[['total_bill','tip','size']].corr()
```

```python
sns.heatmap(df[['total_bill','tip','size']].corr())
```

## Join Plot
##### A join plot allows to study the relationship between 2 numeric variables. The central chart display their correlation. It is usullay a scatterplot, a hexbin  plot, a 2D histogram or a 2D density plot

```python
sns.jointplot(x='tip', y = 'total_bill', data = df, kind = 'hex')
```

```python
sns.jointplot(x='tip', y='total_bill', data=df, kind='reg')
```

### Piar plot
#### A 'pairs plot' is also know as a scatterplot, in which one variable in the same data row is matched with another variable's value, like this: Pairs plots are just elaborations on this, showing all variables paired with all the other variables

```python
sns.pairplot(df)
```

```python
sns.pairplot(df, hue = 'sex')
```

```python
sns.pairplot(df, hue = 'smoker')
```

```python
df['smoker'].value_counts()
```

### Dist plot
#### Dist plot helps us to check the distribution of the columns feature

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'], kde=False, bins=10)
# kde -> kernel density estimation
```

## Categorical Plots
- boxplot
- violinplot
- countplot
- barplot

```python
## Count plot
sns.countplot(x='sex', hue='smoker', data=df, palette='Purples')
# sns.countplot('sex', data=df)
```

```python
sns.countplot(x='day', data = df, palette='Set1')
```

```python
sns.barplot(x='total_bill', y='sex', data=df, palette='Set3')
```

```python
sns.barplot(y='total_bill', x='sex', data=df, palette='Set3')
```

### Box Plot
#### A box and whisker plot(sometimes called a boxplot) is a graph that presents information from a five-number summary

```python
sns.boxplot(x='smoker', y='total_bill', data = df, palette='Set2')
```

```python
sns.boxplot(x='sex', y='total_bill', data = df, palette='Set1')
```

```python
sns.boxplot(x='day', y='total_bill', data = df, palette='rainbow')
```

```python
sns.boxplot(data = df, palette='rainbow')
```

```python
sns.boxplot(x='total_bill', y = 'day', hue = 'smoker', data = df)
```

### Violin Plot
##### Violin plot helps us to see both the distribution of data in terms of kernel density estimation and the box plot

```python
sns.violinplot(x='total_bill', y='day', data=df, palette='rainbow')
```
