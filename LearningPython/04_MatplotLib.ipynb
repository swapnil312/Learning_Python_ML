# MatplotLib Turorial
#### Matplotlib is a plotting library for the Python programming language and its numerical mathematics extension NumPy. It provides an object-oriented API for embedding plots into applications using general-purpose GUI toolkits like Tkinter, wxPython, Qt or GTK+
#### Some of the major pros of matplotlib are
- Generally easy to get started for simple plots
- Support for custom labels and texts
- Great control of every element in a figure
- High-quality output in many formats
- Very customizable in general

```python
import matplotlib.pyplot as plt
%matplotlib inline
# The above line is written to prevent writing of plt.show()
```

```python
import numpy as np
```

```python
## Simple Example

x = np.arange(0,10)
y = np.arange(11,21)
```

```python
plt.scatter(x,y,c='b')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graph in 2D')
plt.savefig('Data/test.jpg')
```

```python
y = x*x
```

```python
# plt.plot(x,y,'go--')
plt.plot(x,y,'r*', linestyle='dashed', linewidth=1, markersize=10)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('2d Diagram')
#use shift tab to know more while cursorjust before braces
# change value of each attribute and see what happens to know better
```

```python
# Creating subplots
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r')
```

```python
plt.subplot(2,2,2)
plt.plot(x,y,'g')
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'go')
plt.subplot(2,2,3)
plt.plot(x,y,'b*--')
plt.subplot(2,2,4)
plt.plot(x,y,'yo-')
```

```python
x = np.arange(-100,100)
y = x**3
plt.plot(x,y)
plt.title("Y = X^3")
plt.xlabel('X axis')
plt.ylabel('Y axis')
```

```python
x = [1, 2, 3]
y = np.array([[1, 2], [3, 4], [5, 6]])
plt.plot(x, y, 'g')
```

```python
# Compute the x and y coordinates for points on a sine curve
x = np.arange(-(2*np.pi),2*np.pi,0.1)
y = np.sin(x)
plt.plot(x,y)
plt.title('Sine Curve')
plt.xlabel('x --->')
plt.ylabel('y --->')
```

```python
x = np.arange(-(2*np.pi),2*np.pi,0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.subplot(2,1,1)
plt.plot(x,y_sin,'g--')
plt.title('Sine')

plt.subplot(2,1,2)
plt.plot(x,y_cos,'b--')
plt.title('Cosine')
```

```python
x = [2,4,6,8]
y = [4,16,36,64]

x2 = [3,9,11]
y2 = [6,18,22]

plt.bar(x, y, align = 'center')
plt.bar(x2, y2, color = 'g')
plt.title('Bar Graph')
plt.xlabel('X axis')
plt.ylabel('Y axis')

plt.show()
```

## Histograms

```python
a = np.array([1,2,3,3,4,4,4,5,5,5,5,6,6,6,6,7,7,7,8,8,9,10])
plt.hist(a, bins = 5) # divide the array into (bin) number of divisions
plt.title('Histogram')
plt.xlabel('X axis')
plt.ylabel('Y axis')
```

## Box Plot using Matplotlib

```python
data = [np.random.normal(0, std, 100) for std in range(1, 4)]
plt.boxplot(data, vert=True, patch_artist=True)
```

## Pie Chart

```python
# Data to plot
labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [215,130,245,210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.3, 0, 0.4, 0) # explode first slice

plt.pie(sizes, explode = explode, labels = labels, colors = colors, autopct = '%1.1f%%', shadow = True)

plt.axis('equal')
plt.show()
```
